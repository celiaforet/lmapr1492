# Etude de l’effet de 3 éléments de symétrie différents chacun pour 1 atome différent

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifParser
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.core.operations import SymmOp
from IPython.display import Image
with MPRester("Jqka7RaOKJvTC6iABC") as me:    
    
    # Téléchargement du fichier CIF
    # Structure for material id
    pars = CifParser("mp9368.cif")
    struct = pars.get_structures()[0]
```

## Symétrie 1 - réflexion par rapport à un plan miroir

```python
sym1 = SymmOp.from_xyz_string('-y,-x,z')
sym1_op = sym1.operate([0.67,0.33,0.33])
im1 = Image(filename='sym1.jpg')
    
print("L’élément de symétrie n°8 du fichier CIF est un plan miroir perpendiculaire à la direction [-1 -1 0]. Cet élément de symétrie envoie tout atome situé en (x,y,z) sur un atome équivalent situé en (-y,-x,z). La matrice qui lui est associée est :")
print(sym1.rotation_matrix)
print("Si on applique l'opération de symétrie à l'atome Rb situé en (2/3,1/3,1/3), il aura pour nouvelle position : ")
print(sym1_op)
print("La figure suivante montre cette application :")
display(im1)
```

## Symétrie 2 - rotation de 120° 

```python
sym2 = SymmOp.from_xyz_string('-y,x-y,z')
sym2_op = sym2.operate([1.00,1.00,1.00])
im2 = Image(filename='sym2.jpg')
print("L’élément de symétrie n°3 du fichier CIF est un axe de rotation d’ordre 2 dans la direction [0 0 1], correspondant à l’arête c de la maille. Cet élément de symétrie envoie tout atome situé en (x,y,z) sur un atome équivalent situé en (-y,x-y,z). La matrice qui lui est associée est :")
print(sym2.rotation_matrix)
print("Si on applique cette opération de symétrie à l'atome Rb situé en (1,1,1), il aura pour nouvelle position : ")
print(sym2_op)
print("La figure suivante montre cette application :")
display(im2)
```

## Symétrie 3 - rotation de 180°

```python
sym3 = SymmOp.from_xyz_string('y,x,-z')
sym3_op = sym3.operate([0.00,1.00,0.23])
im3 = Image(filename='sym3.jpg')
print("L’élément de symétrie n°7 du fichier CIF est un axe de rotation d’ordre 3 dans la direction [1 1 0]. Cet élément de symétrie envoie tout atome situé en (x,y,z) sur un atome équivalent situé en (y,x,-z). La matrice qui lui est associée est :")
print(sym3.rotation_matrix)
print("Si on applique cette opération de symétrie à l'atome Rb situé en (0,1,0.23), il aura pour nouvelle position : ")
print(sym3_op)
print("La figure suivante montre cette application :")
display(im3)
```
